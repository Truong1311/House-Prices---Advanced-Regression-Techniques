```python
class Cat_Coll_Odinal(BaseEstimator, TransformerMixin):
    def __init__(self,cat_feats_ordinal):
        self.cat_feats_ordinal = cat_feats_ordinal
    def fit(self,data):   
        return self
    def transform(self,data):
        df = data[self.cat_feats_ordinal].copy()
        for ft in cat_feats_ordinal:
            df[ft] = df[ft].fillna('None')
            key = df[ft].value_counts().keys()
            value = [i for i in range(len(key))]
            df[ft] = df[ft].replace(to_replace = key, value = value)
        return df
```
